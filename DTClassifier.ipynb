```python
#UAS Data Mining - Decision Tree Classification - Khalifah Imam Ali - M0519054
#import
import pandas as pd #Untuk input output dataset
import numpy as np #NumPy memberikan dukungan untuk himpunan dan matriks

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


#input dataset with pandas -- edit
df = pd.read_csv('https://raw.githubusercontent.com/creativelogy/UAS-Datmin/main/Dataset-Panleukopenia-AI.csv')
dfTest = pd.read_csv('https://raw.githubusercontent.com/creativelogy/UAS-Datmin/main/DataTest-Panleukopenia-AI.csv')

#Cek dataset
df.columns

#Lihat datasetnya ada yang null atau tidak (?)
df.head()

x_data = df.drop(['Reaksi'], axis = 1)
y_data = df['Reaksi']

x_dataTest = dfTest.drop(['Reaksi'], axis = 1)
y_dataTest = dfTest['Reaksi']

model = DecisionTreeClassifier(criterion='gini')
model.fit(x_data, y_data)
model.score(x_data, y_data)
predicted= model.predict(x_dataTest)
	
accuracyTree = accuracy_score(predicted, y_dataTest) * 100
print('Akurasi pada Algoritma Decision Tree menggunakan Hold-out dengan persentase 70:30 = %s' % accuracyTree)


#Visualisasi data
plt.figure(figsize=(20,20))
dec_tree = plot_tree(decision_tree=model, feature_names = df.columns, 
                     class_names =["Positive", "Negative"] , filled = True , precision = 4, rounded = True)
```
